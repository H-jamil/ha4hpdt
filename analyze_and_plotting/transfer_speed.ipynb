```python
import os
import re

# Define the directory
directory = "/home/beams/MJAMIL/hpdt4ha/tests/perf_tools/logs/transfer/send_OS_receive_OS/receiver_throughput/"

# Initialize empty lists
OS_throughput = []
filename_OS_throughput=[]

# Iterate over all the files in the directory
for filename in os.listdir(directory):
    # Open the file
    with open(os.path.join(directory, filename), 'r') as file:
        # Read the lines of the file
        lines = file.readlines()
        mean_throughput_mbps=[]
        for line in lines:
            if "mean throughput" in line:
                # Split the line into words
                words = line.split()
                # The last word in the line is the numerical value we want
                throughput = words[-2]
#                 print(throughput)
                # Check if the value is in Mb/s
                if "Mb/s" in line:
                    # Convert the string to a float and assign it to a variable
                    mean_throughput_mbps.append(float(throughput)/1000)
    OS_throughput.append(sum(mean_throughput_mbps) / len(mean_throughput_mbps))
    filename_OS_throughput.append(filename)
# OS_throughputughputughputughputughputroughput=[47.716422, 65.829876, 89.197698, 101.213129]
print("OS_throughput: ",OS_throughput)
print("filename_OS_throughput", filename_OS_throughput)
```
