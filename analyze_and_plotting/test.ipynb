```python
import re

# Define a list to hold the data of the last two timestamps
timestamps_data = []

with open("test.log", "r") as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        # Check if the line is a timestamp (it does not contain '%' character)
        if '%' not in line:
            # If the list already contains two timestamps, remove the first one
            if len(timestamps_data) == 2:
                timestamps_data.pop(0)

            # Add a new timestamp data
            timestamp = line.strip()
            timestamp_data = {'timestamp': timestamp, 'cpu_usages': []}
            timestamps_data.append(timestamp_data)

        # Check if the line contains cpu usage data (it contains '%' character)
        elif '%' in line:
            cpu_core = re.search(r'%Cpu(\d+)', line).group(1)
            cpu_usage = re.search(r': ([0-9]*\.?[0-9]*)', line).group(1)
            timestamps_data[-1]['cpu_usages'].append({'core': cpu_core, 'usage': float(cpu_usage)})

# Compute the average CPU usage for each timestamp
for timestamp_data in timestamps_data:
    timestamp = timestamp_data['timestamp']
    print(f'Timestamp: {timestamp}')
    for cpu_usage_data in timestamp_data['cpu_usages']:
        core = cpu_usage_data['core']
        usage = cpu_usage_data['usage']
        print(f'Core: {core}, Usage: {usage}')
```
