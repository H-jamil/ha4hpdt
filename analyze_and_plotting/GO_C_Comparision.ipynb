```python
import matplotlib.pyplot as plt
import numpy as np

# sample data
categories = ['Single_Stream_max', 'Single_Stream_average']
variable1 = [0.14986, 0.082]
variable2 = [28.414872, 19.22]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, variable1, width, label='GO')
rects2 = ax.bar(x + width/2, variable2, width, label='C(0MQ)')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Categories')
ax.set_ylabel('Gbps')
ax.set_title('1MB chunk transfer between lynx and polaris using one stream')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# To make it easier to read, you can add a horizontal grid:
ax.grid(axis='y')

# Function to autolabel the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.yscale("log")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# sample data
categories = ['computation_transfer_max', 'computation_transfer_average']
variable1 = [0.120746*1000, 0.078*1000]
variable2 = [0.002633*1000, 0.0023*1000]
variable3 = [0.580437*1000, 0.56841*1000]

x = np.arange(len(categories))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, variable1, width, label='GO')
rects2 = ax.bar(x, variable2, width, label='C(Zlib-0MQ)')
rects3 = ax.bar(x + width, variable3, width, label='C(lz4-0MQ)') # Added third bar


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Categories')
ax.set_ylabel('Mbps')
ax.set_title('1MB chunk pipelined computation and then transfer between lynx and polaris using one stream')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# To make it easier to read, you can add a horizontal grid:
ax.grid(axis='y')

# Function to autolabel the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
fig.tight_layout()
# plt.yscale("log")
plt.show()
```
